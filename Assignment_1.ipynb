```python
#Scatter plots using Matplotlib
```

```python
plt.scatter(iris['SL'],iris['SW'],s=50,color='red')
plt.title('Sepal length vs Sepal Width',fontsize=16)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
plt.scatter(iris['PL'],iris['PW'],s=50,color='GREEN')
plt.title('Petal length vs Petal Width',fontsize=16)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
# Pair plot of iris data using Seaborn
```

```python
sns.pairplot(iris)
```

```python
#Boxplots using Seaborn
```

```python
sns.boxplot(x='Classification',y='SL',data=iris).set_title('Species vs Sepal Length')
plt.show()
```

```python
sns.boxplot(x='Classification',y='SW',data=iris).set_title('Species vs Sepal Width')
plt.show()
```

```python
sns.boxplot(x='Classification',y='PL',data=iris).set_title('Species vs Petal Length')
plt.show()
```

```python
sns.boxplot(x='Classification',y='PW',data=iris).set_title('Species vs Petal Width')
plt.show()
```
